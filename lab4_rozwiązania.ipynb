Wczytaj ramkę danych `tips` korzystając z polecenia:

import seaborn as sns

a następnie odpowiedz na poniższe pytania:

- **(a)** Czy na podstawie danych możemy stwierdzić, że kobiety udzielają wyższych napiwków?
- **(b)** Czy na podstawie danych możemy stwierdzić, że wysokość rachunku opłacanego przez
  palących mężczyzn zależy od dnia, w którym odwiedzili restaurację?


### (a) Czy kobiety udzielają wyższych napiwków?

````python
# Upewniamy się, że kolumna tip_fraction istnieje
if 'tip_fraction' not in tips.columns:
    tips['tip_fraction'] = tips['tip'] / tips['total_bill']

# Porównanie napiwków (wartość bezwzględna i frakcja) wg płci
summary_sex = tips.groupby('sex')[['tip', 'tip_fraction']].agg(['mean', 'median'])
print(summary_sex)
````

````python
# Rozkłady napiwków wg płci
tips.groupby('sex')['tip'].describe()
````

````python
# Test Manna-Whitneya (nieparametryczny) — H0: rozkłady napiwków są identyczne
from scipy import stats

male_tips   = tips.loc[tips['sex'] == 'Male',   'tip']
female_tips = tips.loc[tips['sex'] == 'Female', 'tip']

stat, p_value = stats.mannwhitneyu(female_tips, male_tips, alternative='two-sided')
print(f"Statystyka U = {stat:.1f},  p-value = {p_value:.4f}")
print()
if p_value < 0.05:
    print("Wynik: różnica istotna statystycznie (α=0.05).")
    if female_tips.median() > male_tips.median():
        print("Kobiety udzielają statystycznie wyższych napiwków (mediana).")
    else:
        print("Mężczyźni udzielają statystycznie wyższych napiwków (mediana).")
else:
    print("Wynik: brak podstaw do odrzucenia H0 — różnica NIE jest istotna statystycznie.")
````

### (b) Czy rachunek palących mężczyzn zależy od dnia wizyty?

````python
# Podzbiór: palący mężczyźni
male_smokers = tips.loc[(tips['sex'] == 'Male') & (tips['smoker'] == 'Yes')]
print(f"Liczba obserwacji: {len(male_smokers)}")
male_smokers.groupby('day')['total_bill'].describe()
````

````python
# Test Kruskala-Wallisa — H0: rozkłady rachunków są identyczne we wszystkich dniach
groups = [grp['total_bill'].values for _, grp in male_smokers.groupby('day')]

stat_kw, p_kw = stats.kruskal(*groups)
print(f"Statystyka H = {stat_kw:.3f},  p-value = {p_kw:.4f}")
print()
if p_kw < 0.05:
    print("Wynik: różnica istotna statystycznie — wysokość rachunku zależy od dnia wizyty.")
else:
    print("Wynik: brak podstaw do odrzucenia H0 — nie stwierdzono zależności od dnia.")
````

````python
from nycflights13 import flights, airports, planes, airlines, weather

datasets = {
    'flights'  : flights,
    'airports' : airports,
    'planes'   : planes,
    'airlines' : airlines,
    'weather'  : weather,
}

for name, df in datasets.items():
    print(f"{'='*55}")
    print(f"  {name.upper()}  —  shape: {df.shape}")
    print(f"{'='*55}")
    df.info()
    print()
````

Korzystamy z ramki `planes` z bazy `nycflights13`.

- **(a)** Wybierz tylko unikatowe typy silników (kolumna `engine`).
- **(b)** Jakie typy samolotów były produkowane przez poszczególnych producentów?
  (kolumny `type`, `manufacturer`) — tylko unikalne kombinacje.
- **(c)** Podaj liczbę wystąpień poszczególnych typów silnika (kolumna `engine`).
- **(d)** Policz ile jest unikatowych par (typ silnika, typ samolotu) w danych.
- **(e)** W którym roku dany producent zaczął produkować samoloty o poszczególnych typach silnika?


````python
planes.head(3)
````

### (a) Unikatowe typy silników

````python
unique_engines = planes['engine'].unique()
print(f"Liczba unikalnych typów silników: {len(unique_engines)}")
print(unique_engines)
````

### (b) Typy samolotów wg producentów — unikalne kombinacje

````python
type_by_manufacturer = (
    planes[['manufacturer', 'type']]
    .drop_duplicates()
    .sort_values(['manufacturer', 'type'])
    .reset_index(drop=True)
)
print(f"Liczba unikalnych kombinacji: {len(type_by_manufacturer)}")
type_by_manufacturer
````

### (c) Liczba wystąpień poszczególnych typów silnika

````python
engine_counts = (
    planes['engine']
    .value_counts()
    .rename_axis('engine')
    .reset_index(name='count')
)
engine_counts
````

### (d) Liczba unikatowych par (typ silnika, typ samolotu)

````python
n_unique_pairs = planes[['engine', 'type']].drop_duplicates().shape[0]
print(f"Liczba unikatowych par (engine, type): {n_unique_pairs}")
planes[['engine', 'type']].drop_duplicates().sort_values(['engine', 'type'])
````

### (e) Rok, w którym producent zaczął produkować samoloty danego typu silnika

````python
first_year = (
    planes.dropna(subset=['year'])
    .groupby(['manufacturer', 'engine'])['year']
    .min()
    .reset_index()
    .rename(columns={'year': 'first_year'})
    .sort_values(['manufacturer', 'engine'])
    .reset_index(drop=True)
)
first_year
````
