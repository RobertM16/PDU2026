## Zadanie 1.6[MG] 

Napisz funkcje, ktora wyznacza modę z danego obiektu iterowalnego przechowujacego
obiekty niezmienne jako elementy. Jesli moda nie jest jednoznaczna, zwroć:
- wartość dowolną;
- wartość losowo wybraną.

```python
from collections import Counter
import random

def moda_dowolna(iterable):
    """Zwraca modę (najczęstszy element). Przy remisie — dowolny z najczęstszych."""
    licznik = Counter(iterable)
    return licznik.most_common(1)[0][0]

def moda_losowa(iterable):
    """Zwraca modę. Przy remisie — losowo wybrany spośród najczęstszych."""
    licznik = Counter(iterable)
    max_count = licznik.most_common(1)[0][1]
    najczestsze = [k for k, v in licznik.items() if v == max_count]
    print(najczestsze)
    return random.choice(najczestsze)
```

## Zadanie 1.9[MG] 

Napisz funkcję, która dokonuje standaryzacji wartości wektora numerycznego, tj., dokonuje przekształcenia jego wartości tak by średnia wynikowego wektora była równa 0, a odchylenie standardowe 1 (or. Z-scoring).

```python
def standardize(x):
    """
    Funkcja dokonuje standaryzacji wektora x
    Parameters
    ----------
    x : tablica numpy.ndarray jednowymiarowa  
        (wektor)

    Returns
    -------
    Tablica jednowymiarowa będąca wystandaryzowaną wersją x
    
    """
    
    assert x.ndim == 1 # czy wymiar odpowiedni
    # w przypadku tablic numpy.ndarray zachodzi uzgadnianie kształtów -
    # przez przypadek możemy więc (bez błędu) policzyć coś co nie jest 
    # poprawnym (oczekiwanym) przez nas wynikiem.
    # Przed taką sytuacją się zabezpieczamy prz użyciu assert.
    #
    # Korzystamy z wektoryzacji:
    # 1.) od każdego elementu odejmujemy średnia: x - x.mean()
    # 2.) każdy tak zmieniony element dzielimy przez odchylenie: 
    #     (...) / x.std(ddof = 1)
    # Uwaga:
    # wariancja = (\sum_{i=1}^{n} (x[i] - srednia_x)**2) / (n - ddof)
    # dla ddof = 1 mamy najlepsze własności statystyczne
    # ddof=1 (jak w kodzie) — estymator nieobciążony wariancji, dzieli przez n-1. Używany gdy x to próbka i chcemy estymować odchylenie standardowe populacji.
    # ddof=0 — dzieli przez n. Używany gdy x to cała populacja albo gdy robimy czysto techniczną normalizację (np. preprocessing w ML, batch normalization).
    
    return (x - x.mean()) / x.std(ddof = 1)
```

$$w_i = \sum_{j=1}^{k}\frac{x_{i+j-1}}{k}.$$


```python
import numpy as np
import matplotlib.pyplot as plt

def srednia_ruchoma(x, k):
    """
    Funkcja wyznacza średnią ruchomą dla danego
    wektra x
    
    Więcej szczegółów w pliku .pdf, .html lub .ipynb
    Parameters
    ----------
    x : tablica numpy.ndarray jednowymiarowa  
        (wektor)
    
    Returns
    -------
    Tablica numpy.ndarray jednowymiarowa  
    (wektor)
    """
    # 1.) np.cumsum() - skumulowane sumy
    w = np.cumsum(x)
    # 2.) odejmujemy od sum skumulowanych 
    #     wartości elementów tak by uzyskac tylko
    #     sumy k kolejnych elementów
    w[k:] -= w[:-k]
    # 3.) Wykorzystujemy wektoryzację wektor i skalar
    #     by każdą taka sumę podzielić przez k
    return w[(k-1):]/k

def srednia_ruchoma2(x, k):
    
    return np.convolve(x, np.ones(k), 'valid') / k

def srednia_ruchoma3(x, k):
    
    import pandas as pd

    w = pd.Series(x)

    return w.rolling(k).mean()


# Jako przykład wygenerujemy rysunek:
# Generujemy dane:
x = np.random.randn(100)
y = x.cumsum()
# Funkcja plt.plot() rysuje krzywą łącząc punkty zawarte
# w tablicy y
plt.plot(y, 'k-', label="dane") # 'k-' wygląd krzywej (czarna, ciągła)
k = 5
plt.plot(range(k-1, len(y)), srednia_ruchoma(y, k), 'r--', label="MA(3)") # 'r--` (czerwona, przerywana)
k = 10
plt.plot(range(k-1, len(y)), srednia_ruchoma(y, k), 'b-.', label="MA(5)") # 'b-.' (niebieska, kropka-kreska

plt.legend() # legenda (korzysta z etykiet label)
plt.show()
```

## Zadanie 1.13[MG]

Dokonaj przekształcenia \textit{softmax} każdego wiersza macierzy 
$\mathbf{X}\in\mathbb{R}^{n\times k}$, tzn. przekształcenia postaci: 
$$x_{i,j}\mapsto \frac{\exp(x_{i,j})}{\sum_{l=1}^k \exp(x_{i,l})}.$$

Następnie dokonaj odkodowania każdego wiersza (*one-hot decode*), 
tj. dla każdego wiersza należy znaleźć numer kolumny o wartości najbardziej 
zbliżonej do $1$. Zwróć wektor $n$-elementowy.


```python
def softmax1(X):
    r"""
    Softmax transformation
    i.e.
    x_{ij} -> ( exp(x_{ij}) / \sum_{l=1}^{k} exp_{x_{il}} )
    
    Parameters
    ----------
    X : matrix (two dimensional numpy.ndarray array) 

    Returns
    -------
    Function return transformed matrix of size nxk (numpy.ndarray)
    """    
    import scipy.special 
    
    return scipy.special.softmax(X, axis=1)
```

```python
def softmax2(X):
    r"""
    Softmax transformation
    i.e.
    x_{ij} -> ( exp(x_{ij}) / \sum_{l=1}^{k} exp_{x_{il}} )
    
    Parameters
    ----------
    X : matrix (two dimensional numpy.ndarray array) 

    Returns
    -------
    Function return transformed matrix of size nxk (numpy.ndarray)
    """    
    Y = np.exp(X) # funkcja zwektoryzowana
    # 1.) axis - po której będziemy liczyć sumę
    # tu 1 czyli dostaniemy wartości sumy w każdym wierszu
    # liczymy po kolumnach
    # 2.) (-1, 1) automatyczne dostosowanie liczby wierszy, 1 kolumna
    # dzieki temu dostaniemy wektoryzację po kolumnach
    return Y / Y.sum(axis = 1).reshape(-1, 1) 
```
